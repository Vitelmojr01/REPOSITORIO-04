Resolvendo Exercicio 01 - item 01

```python
import pandas as pd, numpy as np, statsmodels.formula.api as smf
df = pd.read_csv('previsao_de_renda.csv')
print("Base carregada:", df.shape)

df = df.drop(columns=['Unnamed: 0','data_ref','id_cliente'])
print("Colunas após limpeza:", df.columns.tolist())

modelo1 = smf.ols("""np.log(renda) ~ posse_de_veiculo + posse_de_imovel + qtd_filhos + idade +
tempo_emprego + qt_pessoas_residencia +
C(sexo, Treatment(reference='F')) +
C(tipo_renda, Treatment(reference='Assalariado')) +
C(educacao, Treatment(reference='Secundário')) +
C(estado_civil, Treatment(reference='Casado')) +
C(tipo_residencia, Treatment(reference='Casa'))""", data=df).fit()

print("R² ajustado:", modelo1.rsquared_adj)
print("AIC:", modelo1.aic)
print(modelo1.summary())
```

Resolvendo tarefa 02 - modulo 13 - exercicio 01 - item 02
Mantenha sempre a categoria mais frequente como casela de referência

```python
modelo2 = smf.ols("""np.log(renda) ~ posse_de_veiculo + posse_de_imovel + qtd_filhos + idade +
tempo_emprego + qt_pessoas_residencia +
C(sexo, Treatment(reference='F')) +
C(tipo_renda, Treatment(reference='Assalariado')) +
C(educacao, Treatment(reference='Secundário')) +
C(estado_civil, Treatment(reference='Casado'))""", data=df).fit()

print("Modelo 1 AIC:", modelo1.aic, "| Modelo 2 AIC:", modelo2.aic)
print("Modelo 1 R²aj:", modelo1.rsquared_adj, "| Modelo 2 R²aj:", modelo2.rsquared_adj)
```

item 03 - Avaliação: As notas do modelo indicam possível multicolinearidade devido ao condition number alto. Isso sugere que algumas variáveis podem estar correlacionadas, o que pode afetar a estabilidade dos coeficientes, mas o modelo ainda pode ser interpretado com cautela.

03-01- Removendo  variável posse de veiculo

```python
modelo3 = smf.ols("""np.log(renda) ~ posse_de_veiculo + posse_de_imovel +
idade + tempo_emprego + qt_pessoas_residencia +
C(sexo, Treatment(reference='F')) +
C(tipo_renda, Treatment(reference='Assalariado')) +
C(estado_civil, Treatment(reference='Casado'))""", data=df).fit()

print("Comparação final:")

print("\nModelo inicial:")
print("AIC:", modelo1.aic)
print("R² ajustado:", modelo1.rsquared_adj)

print("\nModelo final:")
print("AIC:", modelo3.aic)
print("R² ajustado:", modelo3.rsquared_adj)

print("\nResumo modelo final:")
print(modelo3.summary())
```

Algumas categorias pouco significativas de tipo_renda (dependendo da sua versão anterior)
