# Regressão 01 - tarefa 02 - transformações em X e Y

00. Carregando a base

````python
# Importando as Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
from seaborn import load_dataset

import numpy as np

import statsmodels.formula.api as smf
````

````python
# A base dados será uma base da própria bibliotca do seaborn que seria o do problema de gorjetas, como feito em aula. Ou seja

tips = sns.load_dataset("tips")
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips['tip_pct'] = tips['tip'] / tips['net_bill']
tips.drop(['total_bill', 'sex', 'smoker', 'day', 'time', 'size'], axis=1, inplace=True)
tips.head()
````

Criando a variável gorjetas

````python
import pandas as pd, numpy as np, seaborn as sns, statsmodels.formula.api as smf
gorjetas = sns.load_dataset('tips')
gorjetas['net_bill'] = gorjetas['total_bill'] - gorjetas['tip']
gorjetas['tip_pct'] = gorjetas['tip'] / gorjetas['net_bill']
````

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

Resolvendo item 1 - tarefa 02 - modulo 12

````python
#Reproduzir o modelo base
modelo1 = smf.ols('tip_pct ~ net_bill', data=gorjetas).fit()

print(modelo1.summary())
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

Resolvendo item 2 - tarefa 02 - modulo 12

````python
modelo2 = smf.ols('tip_pct ~ np.log(net_bill)', data=gorjetas).fit()

print(modelo2.summary())
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

Resolvendo item 3 - tarefa 02 - modulo 12

````python
modelo3 = smf.ols('tip_pct ~ net_bill + I(net_bill**2)', data=gorjetas).fit()

print(modelo3.summary())
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

Resolvendo item 4 - tarefa 02 - modulo 12

````python
modelo4 = smf.ols('np.log(tip_pct) ~ net_bill', data=gorjetas).fit()

print(modelo4.summary())
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

REsolvendo item 5 - tarefa 02 - modulo 12

````python
print("Resolvendo item 5 - tarefa 02 - modulo 12")

# modelo log-log
modelo5 = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data=gorjetas).fit()

# resumo do modelo
print(modelo5.summary())

# valores preditos no log
pred_log = modelo5.fittedvalues

# voltando para escala original tip_pct
gorjetas['tip_pct_pred'] = np.exp(pred_log)

# modelo auxiliar para calcular R² em tip_pct
modelo5_r2 = smf.ols('tip_pct ~ tip_pct_pred', data=gorjetas).fit()

print("\nR² na escala tip_pct:")
print(modelo5_r2.rsquared)

# gráfico para visualizar
sns.regplot(x='net_bill', y='tip_pct', data=gorjetas)
plt.title("Modelo log-log convertido para tip_pct")
plt.show()
````

###  6. qual modelo resultou no melhor ajuste?

RESPOSTA ITEM 06: O modelo que apresentou melhor ajuste foi log(tip_pct) em função de log(net_bill), pois apresentou o maior R² (≈0.31). Como maior R² indica melhor capacidade explicativa, este modelo foi o que melhor explicou a variabilidade do percentual de gorjeta.
